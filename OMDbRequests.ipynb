```python
#import dependencies
import requests
import json
import pandas as pd
from config import api_key
from pprint import pprint
```

# TNG

```python
#Loop through season numbers to return list of dictionaries for all TNG seasons

seasons = ['1','2','3','4','5','6','7']
url = "http://www.omdbapi.com/?apikey=" + api_key + "&t=Star Trek: The Next Generation&Season="
tng_responses=[]

for entry in seasons:
    #tng_responses = requests.get(url+season).json()
    #print(tng_responses)
    tng_responses.append(requests.get(url+entry).json())
```

```python
#loop through responses to append values to lists

#create empty lists to hold values from loop
season = []
title = []
episode = []
rating = []
series = []

index = 0

for x in tng_responses:
    for response in tng_responses[index]['Episodes']:
        title.append(response['Title'])
        season.append(index+1)
        episode.append(response['Episode'])
        rating.append(response['imdbRating'])
        series.append(tng_responses[0]['Title'])
        
    index = index + 1

title
```

```python
df = pd.DataFrame(list(zip(series, season, episode, title, rating)),
               columns =['Series', 'Season', 'Episode', 'Title', 'IMDb Rating'])
df
```
